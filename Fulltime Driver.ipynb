```python
fulltimedrivershift = []
fulltimedriverduration = []
fulltimedriverstartingtime = []
fulltimedriverstartinglocation = []
for i in range(numfulldriver):
    summaryduration = []
    summarystartingtime = []
    summarystartinglocation = []
    # number of shifts of this driver
    numberofshifts = fulltimedrivernumshift[i][0]
    fulltimedrivershift.append(numberofshifts)
    # estimate primary shift duration
    primarydurationU = fullpr(numberofshifts)
    pdurdraw = np.random.random()
    for j in range(len(primarydurationU)):
        if(j==0):
            if(pdurdraw<primarydurationU[j]):
                primaryduration=1
        else:
            if(pdurdraw<primarydurationU[j] and pdurdraw>=primarydurationU[j-1]):
                primaryduration=j+1
    summaryduration.append(primaryduration)
    if(numberofshifts>=2):
        # estimate second shift duration
        secondarydurationU = fullsec(numberofshifts,primaryduration)
        secondaryduration = 99
        while(secondaryduration>primaryduration):
            sdurdraw = np.random.random()
            for j in range(len(secondarydurationU)):
                if(j==0):
                    if(sdurdraw<secondarydurationU[j]):
                        secondaryduration=1
                else:
                    if(sdurdraw<secondarydurationU[j] and sdurdraw>=secondarydurationU[j-1]):
                        secondaryduration=j+1
        summaryduration.append(secondaryduration)
        if(numberofshifts>=3):
            #estimate tertiary shift duration
            tertiarydurationU = fullter(primaryduration,secondaryduration)
            tertiaryduration = 99
            while(tertiaryduration>secondaryduration):
                tdurdraw = np.random.random()
                for j in range(len(tertiarydurationU)):
                    if(j==0):
                        if(tdurdraw<tertiarydurationU[j]):
                            tertiaryduration=1
                    else:
                        if(tdurdraw<tertiarydurationU[j] and tdurdraw>=tertiarydurationU[j-1]):
                            tertiaryduration=j+1
            summaryduration.append(tertiaryduration)
    #estimate primary starting time
    primarytimeU = fullps(numberofshifts,primaryduration)
    ptimedraw = np.random.random()
    for j in range(len(primarytimeU)):
        if(j==0):
            if(ptimedraw<primarytimeU[j]):
                primarytime=1
        else:
            if(ptimedraw<primarytimeU[j] and ptimedraw>=primarytimeU[j-1]):
                primarytime=j+1
    summarystartingtime.append(primarytime)
    #estimate secondary starting time
    if(numberofshifts>=2):
        secondarytimeU, secondarytimeP = fullss(numberofshifts,secondaryduration,primarytime,primaryduration)
        sectimedraw = np.random.random()
        secondarytime = primarytime
        ava = list(range(1,25))
        for k in range(primaryduration):
            if(primarytime+k<=24):
                ava.remove(primarytime+k)
            else:
                ava.remove(primarytime+k-24)
        for k in range(secondaryduration):
            if(primarytime-k<0 and primarytime-k in ava):
                ava.remove(primarytime-k+24)
            else:
                if(primarytime-k-1 in ava):
                    ava.remove(primarytime-k-1)
        tempsecondarytimeP = []
        for j in ava:
            tempsecondarytimeP.append(secondarytimeP[j-1])
        newsecondarytimeU = []
        tempsumU = 0
        for j in tempsecondarytimeP:
            tempsumU+=j/sum(tempsecondarytimeP)
            newsecondarytimeU.append(tempsumU)
        for j in range(len(newsecondarytimeU)):
            if(j==0):
                if(sectimedraw<newsecondarytimeU[j]):
                    tempsecondarytime=1
            else:
                if(sectimedraw<newsecondarytimeU[j] and sectimedraw>=newsecondarytimeU[j-1]):
                    tempsecondarytime=j+1
        secondarytime = ava[tempsecondarytime-1]
        summarystartingtime.append(secondarytime)
        if(numberofshifts>=3):
            tertiarytimeU, tertiarytimeP = fullts(numberofshifts,tertiaryduration,secondaryduration,secondarytime,primarytime,primaryduration)
            tertimedraw = np.random.random()
            tertiarytime = primarytime
            tava = list(range(1,25))
            for k in range(primaryduration):
                if(primarytime+k<=24):
                    tava.remove(primarytime+k)
                else:
                    tava.remove(primarytime+k-24)
            for k in range(tertiaryduration):
                if(primarytime-k<0 and primarytime-k+24 in tava):
                    tava.remove(primarytime-k+24)
                else:
                    if(primarytime-k-1 in tava):
                        tava.remove(primarytime-k-1)
            for k in range(secondaryduration):
                if(secondarytime+k<=24 and secondarytime+k in tava):
                    tava.remove(secondarytime+k)
                elif(secondarytime+k>24 and secondarytime+k-24 in tava):
                    tava.remove(secondarytime+k-24)
            for k in range(tertiaryduration):
                if(secondarytime-k<0 and secondarytime-k+24 in tava):
                    tava.remove(secondarytime-k+24)
                elif(secondarytime-k>=0 and secondarytime-k in tava):
                    tava.remove(secondarytime-k)
            temptertiarytimeP = []
            for j in tava:
                temptertiarytimeP.append(tertiarytimeP[j-1])
            newtertiarytimeU = []
            temptsumU = 0
            for j in temptertiarytimeP:
                temptsumU+=j/sum(temptertiarytimeP)
                newtertiarytimeU.append(temptsumU)
            for j in range(len(newtertiarytimeU)):
                if(j==0):
                    if(tertimedraw<newtertiarytimeU[j]):
                        temptertiarytime=1
                else:
                    if(tertimedraw<newtertiarytimeU[j] and tertimedraw>=newtertiarytimeU[j-1]):
                        temptertiarytime=j+1
            tertiarytime = ava[temptertiarytime-1]
            summarystartingtime.append(tertiarytime)
    # estimate first starting location
    primarylocationU = fullpsl()
    plocdraw = np.random.random()
    for j in range(len(primarylocationU)):
        if(j==0):
            if(plocdraw<primarylocationU[j]):
                primarylocation=1
        else:
            if(plocdraw<primarylocationU[j] and plocdraw>=primarylocationU[j-1]):
                primarylocation=j+1
    summarystartinglocation.append(primarylocation)
    if(numberofshifts>=2):
        secondarylocationU = fullssl()
        slocdraw = np.random.random()
        for j in range(len(secondarylocationU)):
            if(j==0):
                if(slocdraw<secondarylocationU[j]):
                    secondarylocation=1
            else:
                if(slocdraw<secondarylocationU[j] and slocdraw>=secondarylocationU[j-1]):
                    secondarylocation=j+1
        summarystartinglocation.append(secondarylocation)
        if(numberofshifts>=3):
            tertiarylocationU = fulltsl()
            tlocdraw = np.random.random()
            for j in range(len(tertiarylocationU)):
                if(j==0):
                    if(tlocdraw<tertiarylocationU[j]):
                        tertiarylocation=1
                else:
                    if(tlocdraw<tertiarylocationU[j] and tlocdraw>=tertiarylocationU[j-1]):
                        tertiarylocation=j+1
            summarystartinglocation.append(tertiarylocation)
    fulltimedriverduration.append(summaryduration)
    fulltimedriverstartingtime.append(summarystartingtime)
    fulltimedriverstartinglocation.append(summarystartinglocation)
```
